```python
#importing libs
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#download dataset and setting "Txn_date" as index
df = pd.read_csv("data_no_outliers.csv")
df = df.set_index('Txn_date')
df.index = pd.to_datetime(df.index)
df = df.sort_index()
df
```

```python
#df.sample swarmplot
sample_data = df.sample(frac=0.1, random_state=1)

plt.figure(figsize=(10, 6))
sns.swarmplot(x=sample_data['Product_Category'], y=sample_data['Total_Amt'])
plt.xticks(rotation=-45)
plt.xlabel('Product Category')
plt.ylabel('Total Sales ($)')
plt.title('Sales by Product Category')
plt.show()
```

```python
sample_data = df.sample(frac=0.1, random_state=1)

plt.figure(figsize=(10, 6))
sns.boxplot(x=sample_data['Product_Category'], y=sample_data['Total_Amt'])
plt.xticks(rotation=-45)
plt.xlabel('Product Category')
plt.ylabel('Total Sales ($)')
plt.title('Sales by Product Category')
plt.show()
```
