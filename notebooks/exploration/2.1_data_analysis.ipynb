```python
#importing libs
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#download dataset and setting "Txn_date" as index
df = pd.read_csv("data_no_outliers.csv")
df = df.set_index('Txn_date')
df.index = pd.to_datetime(df.index)
df = df.sort_index()
df
```

```python
df.info()
```

```python
total_sales = df['Total_Amt'].sum()
total_sales
```

```python
#data analysis
#identifying 'unique_customer_number'

unique_customer_number = df['Cust_id'].nunique()
unique_customer_number
```

```python
customer_freq = df['Cust_id'].value_counts()
customer_freq.describe()
```

```python
#how old is this store?

age = df.index.max() - df.index.min()
age
```

```python
#let's build time based DF's

df['Year'] = df.index.year
df['Month'] = df.index.month
df['Day'] = df.index.day
#day of the week and weekends
df['DayOfWeek'] = df.index.dayofweek
df['Weekend'] = df['DayOfWeek'].apply(lambda x: 1 if x >= 5 else 0)
```

```python
#Plot total monthly sales by year 
plt.figure(figsize=(12,6))
df.groupby(['Year', 'Month'])['Total_Amt'].sum().plot(kind='line', title='Monthly sales by year', ylabel = 'Total Sales ($ in Millions)')

plt.show()
```

```python
#Plot total monthly sales 
plt.figure(figsize=(15,6))
df.groupby('Month')['Total_Amt'].sum().plot(kind='bar', title='Monthly sales', xlabel='Month (January - December)', ylabel = 'Total Sales ($ in Millions)')

plt.show()
```

```python
#pivot plot All three Years monthly Sales on one plot

pivot_sales = pd.pivot_table(df, values='Total_Amt', index='Month', columns='Year', aggfunc='sum')
pivot_sales.plot(figsize=(15,6),kind='bar', title="Monthly Sales by Year", 
                 xlabel='Month (January - December)', 
                 ylabel='Total Sales ($ in Millions)')
```

```python
#pivot plot All three Years daily Sales on one plot

pivot_sales = pd.pivot_table(df, values='Total_Amt', index='DayOfWeek', columns='Year', aggfunc='sum')
pivot_sales.plot(figsize=(15,6),kind='bar', title="Daily Sales by Year", 
                 xlabel='Days of the Week (Monday - Sunday)', 
                 ylabel='Total Sales ($ in Millions)')
```

```python
#pivot plot All three Years monthly Sales on one plot

pivot_sales = pd.pivot_table(df, values='Total_Amt', index='Weekend', columns='Year', aggfunc='sum')
pivot_sales.plot(figsize=(15,6),kind='bar', title="Weekend Sales by Year", 
                 xlabel='Weekends (Saturday - Sunday)', 
                 ylabel='Total Sales ($ in Millions)')
```

```python
sales = df.groupby('Product_Category')['Total_Amt'].sum()


plt.figure(figsize=(10, 6))
sns.barplot(x=sales.index, y=sales.values, palette="coolwarm")
plt.xticks(rotation=-45)
plt.xlabel('Product Category')
plt.ylabel('Total Sales ($ in Millions)')
plt.title('Sales by Product Category')
plt.show()
```

```python
#Plot in boxplot sales by category
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Product_Category'], y=df['Total_Amt'])
plt.xticks(rotation=-45)
plt.xlabel('Product Category')
plt.ylabel('Total Sales ($)')
plt.title('Sales by Product Category')
plt.show()
```

```python
#Plot in boxplot sales by Sub_category
plt.figure(figsize=(15, 6))
sns.boxplot(x=df['Product_Sub_category'], y=df['Total_Amt'])
plt.xticks(rotation=-45)
plt.xlabel('Product Subcategory')
plt.ylabel('Total Sales ($)')
plt.title('Sales by Product Subcategory')
plt.show()
```
