```python
#importing libs
import pandas as pd
```

```python
#download raw data, set as DF
excel = pd.ExcelFile('SuperMarket_Transaction_Data.xlsx')
df = excel.parse(excel.sheet_names[0])
```

```python
#exploring data set
df.head()
```

```python
#checking info
df.info()
```

```python
#checking for duplicates
duplicates = df.duplicated().sum()
print(duplicates)
```

```python
#droping duplicates
df= df.drop_duplicates()
```

```python
#confirming that there is no dublicates
duplicates = df.duplicated().sum()
print(duplicates)
```

```python
#separating returns from DF
df_returns = df[df['Qty'] < 0]

```

```python
#df_returns info
df_returns.info()
```

```python
#DF with no returns
df = df[df['Qty']>=0]
```

```python
df.info()
```

```python
#double check that there is no returns (negative values)
neg_values = (df['Total_Amt'] < 0).any()
neg_values
```

```python
#stat for numeric columns
df[['Qty', 'Rate', 'Tax', 'Total_Amt']].describe()
```

```python
#unique for category types
print(df['Product_Sub_category'].unique())
print(df['Product_Category'].unique())
print(df['Store_type'].unique())
```

```python
#NA values
df.isna().sum()
```

```python
#saving clean data
df.to_csv('clean_data.csv', index=False)
df
```

```python
#saving clean data
df_returns.to_csv('clean_data_returns.csv', index=False)
```

```python
#checking for outliers. q1, q3, iqr

q1 = df['Total_Amt'].quantile(0.25)
q3 = df['Total_Amt'].quantile(0.75)
iqr = q3 - q1

upper = q3+1.5*iqr
lower = q1 - 1.5*iqr

outliers = df[(df['Total_Amt'] < lower) | (df['Total_Amt'] > upper)]

outliers
```

```python
print('q1 - ', q1, ' q3 - ', q3,'upper - ', upper,'lower - ', lower)
```

```python
#DF without outliers
df_no_outliers = df[(df['Total_Amt'] >= lower) & (df['Total_Amt'] <= upper)]
df_no_outliers
```

```python
#saving dataset without outliers
df_no_outliers.to_csv('data_no_outliers.csv', index=False)
```
